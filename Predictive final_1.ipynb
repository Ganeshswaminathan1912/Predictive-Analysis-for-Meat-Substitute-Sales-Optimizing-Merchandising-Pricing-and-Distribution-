Feature Engineering

```python
# Price Elasticity Features
data['Price_Elasticity_Unit'] = data['Price per Unit'] / (data['Unit Sales'] + 1e-5)  # Avoid division by zero
data['Price_Elasticity_Incremental'] = data['Price per Unit'] / (data['Incremental Units'] + 1e-5)

# Weighted Metrics
data['ACV_Weighted_Price_Unit'] = data['ACV Weighted Distribution'] * data['Price per Unit']
data['ACV_Weighted_Price_Volume'] = data['ACV Weighted Distribution'] * data['Price per Volume']

# Volume Ratios
data['Unit_to_Volume_Ratio'] = data['Price per Unit'] / (data['Price per Volume'] + 1e-5)

# Sales Proportion
data['Incremental_to_Total_Sales'] = data['Incremental Units'] / (data['Unit Sales'] + 1e-5)

# Interaction Between Features
data['Price_Distribution_Interaction'] = data['Price per Unit'] * data['ACV Weighted Distribution']
data['Volume_Distribution_Interaction'] = data['Price per Volume'] * data['ACV Weighted Distribution']

# High-Low Sales Ratio
data['High_Low_Sales_Ratio'] = data['High_Low_Sales'] / (data['Unit Sales'] + 1e-5)

# Display the first few rows with the new features
print(data.head())
```

Outliers

```python
numerical_column_names = data.select_dtypes(include=[np.number]).columns
```

```python
# For Z-score
from scipy.stats import zscore
numerical_columns = data.select_dtypes(include=np.number)  # add all numerical columns here
z_scores = np.abs(zscore(numerical_columns))
outliers_z = (z_scores > 4).any(axis=1)

# For IQR
Q1 = numerical_columns.quantile(0.25)
Q3 = numerical_columns.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 2.5 * IQR
upper_bound = Q3 + 2.5 * IQR
outliers_iqr = ((numerical_columns < lower_bound) | (numerical_columns > upper_bound)).any(axis=1)

# Combine both methods
outliers_combined = outliers_z | outliers_iqr
outlier_rows = data[outliers_combined]
```

```python
# Get the rows with outliers
outlier_rows = data[outliers_combined]

# Optionally, you can print the outliers to review
print(outlier_rows)
```

```python
# Calculate correlation matrix for numerical columns
correlation_matrix = log_data.select_dtypes(include=np.number).corr().abs()

# Identify the upper triangle of the correlation matrix
upper_triangle = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))

# Identify columns to drop based on correlation threshold
to_drop = [column for column in upper_triangle.columns if any(upper_triangle[column] > 0.8)]

# Exclude target variables from the columns to drop
target_variables = ['Unit Sales', 'Incremental Units']
to_drop = [column for column in to_drop if column not in target_variables]

# Drop the identified columns
log_data = log_data.drop(columns=to_drop)
```

```python
from sklearn.cluster import KMeans
inertia = []
k_values = range(1,10,1)

for i in k_values:
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
    print("Inertia:", kmeans.inertia_)
```

Graph to see optimal K (Elbow Method)

```python
plt.plot(k_values,inertia)
plt.xlabel('k_values')
plt.ylabel('inertia')
plt.title('Simple Plot')
plt.show()
```
